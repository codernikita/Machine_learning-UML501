```python
from collections import Counter
lst=[1,2,3,2,1,2,4,5,1,1]
counter=Counter(lst)
most_common=counter.most_common(1)[0]
print("most_frequent element:",most_common[0])
print("count:",most_common[1])
indices=[i for i,v in enumerate(lst) if v==most_common[0]]
print("indices:",indices)
```

```python
arr=np.array([1,2,3,2,1,2,4,5,1,1])
values,counts=np.unique(arr,return_counts=True)
max_val=values[np.argmax(counts)]
indices=np.where(arr==max_val)[0]
print(indices)
```

```python
arr=np.array([1,2,3,2,1,2,4,5,1,1])
freq={}
for item in arr:
  freq[item]=freq.get(item,0)+1
  #get will check if item is available give value corresponding to that else return default value 0
most_freq=max(freq,key=freq.get)
print("most frequent element is:",most_freq)
print("occurence :",freq[most_freq])
```

```python
def img_to_array(path):
  img=cv2.imread(path)
  if img is None:
    raise FileNotFoundError(f"image not found at {path}")
  if len(img.shape)==2:
    np.savetxt("image_Array.txt",img,fmt="%d")
    print("saved grayscale image to image_Array.txt")
  else:
    img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    h,w,c=img_rgb.shape
    img_rgb_reshaped=img_rgb.reshape(-1,c)
    np.savetxt("image_array.txt",img_rgb_reshaped,fmt="%d")
    print("saved color image to image_array.txt")
img_to_array("img_test.jpeg")
```
